```python
import matplotlib.pyplot as plt

#custom data:
months=["jan","feb","march","april","may","june","july","august","september","october","november","december"]
sales=[150,180,210,190,230,200,160,220,310,400,550,800]

plt.figure(figsize=(12,6))

#plotting with custom style:
plt.plot(months,sales,marker='s',linestyle='--',color='green',linewidth=2,label='Gadget Sales 2024')

#customization:
plt.title("annual tech sales growth trend",fontsize=14, fontweight='bold',color='darkgreen')
plt.xlabel("months of the year",fontsize=12)
plt.ylabel("revenue(in $1000s)",fontsize=12)
plt.legend()
plt.show()
```
